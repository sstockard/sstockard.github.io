```python
import sqlite3
import pandas as pd
conn = sqlite3.connect("factbook.db")

a = "SELECT * FROM sqlite_master WHERE type='table';"
pd.read_sql_query(a, conn)
```

```python
b = "SELECT * FROM facts;"
pd.read_sql_query(b, conn)
```

```python
c = "SELECT MIN(population), MAX(population), MIN(population_growth), MAX(population_growth) from facts"
pd.read_sql_query(c, conn)
```

```python
d = "SELECT name FROM facts WHERE population = 0;"
pd.read_sql_query(d, conn)
```

```python
f = "SELECT population, population_growth, birth_rate, death_rate FROM facts WHERE population > 0 AND population < 256490011;"
pd.read_sql_query(f, conn)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

data = pd.read_sql_query(f, conn)

fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

data["birth_rate"].hist(ax=ax1)
ax1.set_xlabel("birth_rate")
data["death_rate"].hist(ax=ax2)
ax2.set_xlabel("death_rate")
data["population_growth"].hist(ax=ax3)
ax3.set_xlabel("population_growth")
data["population"].hist(ax=ax4)
ax4.set_xlabel("population")

plt.show()
```
